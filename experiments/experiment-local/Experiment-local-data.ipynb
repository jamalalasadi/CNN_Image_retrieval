```python
import sys
import os
import time
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import matplotlib
import matplotlib as mpl
from matplotlib import cm
blues   = cm.get_cmap(plt.get_cmap('Blues'))
greens  = cm.get_cmap(plt.get_cmap('Greens'))
reds    = cm.get_cmap(plt.get_cmap('Reds'))
oranges = cm.get_cmap(plt.get_cmap('Oranges'))
purples = cm.get_cmap(plt.get_cmap('Purples'))
greys   = cm.get_cmap(plt.get_cmap('Greys'))
```

```python
images = pd.read_csv('/home/usr/local.csv')
fig = plt.figure(figsize=(24, 11))
ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.
font = {'family' : 'normal','size'   : 30}
matplotlib.rc('font', **font)
cnn_time = images['cnn_time']
accuracy = images['accuracy']
ax2.set_ylabel('CNN Prediction Accuracy (%)',fontsize=30)
ax.set_ylabel('CNN Prediction Speed (sec) ',fontsize=30)
ax.set_xlabel('Image',fontsize=30)
degrees = 70
plt.xticks(rotation=degrees)
cnn_time.plot(kind='bar', color=greens(150), ax=ax,width = 0.17, position=0, log =1,legend = False)
accuracy.plot(kind='bar', color=reds(150), ax=ax2,width = 0.17, position=1,legend = False)
ax.bar(images['cnn_time'],images['img_name'])
_ = ax.legend(["CNN Time"],loc='upper center', bbox_to_anchor=(0.35, -0.1),ncol=3,fontsize= 30)
_ = ax2.legend(["CNN Accuracy"],loc='upper center', bbox_to_anchor=(0.57, -0.1),ncol=3,fontsize= 30)
_ = ax.grid(which='both')
```
